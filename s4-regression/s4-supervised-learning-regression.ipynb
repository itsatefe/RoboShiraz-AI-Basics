### Step 3: üß™ Part 1: Linear Regression with 1 Feature (Line)

In this example, we have **one input feature** and **one target**.  
We want to predict a value based on just **one number**.

---

#### 3.1 - ‚úÖ What Does That Mean?

You‚Äôre trying to **learn a relationship** like this:

> üìä ‚ÄúIf I know **X**, can I guess **Y**?‚Äù

For example:

- **X** = Hours a student studies  
- **Y** = Their exam score  

---

#### 3.2 - üìê Mathematically:

$$
y = \beta_0 + \beta_1 x
$$

Where:
- $ x $ is the input (feature)
- $ y $ is the predicted value (target)
- $ \beta_0 $ is the intercept (starting value when x = 0)
- $ \beta_1 $ is the slope (how much y increases for each unit of x)

---

#### 3.3 - üîç What's the Difference Between Intercept and Slope?

| Term       | Meaning |
|------------|---------|
| **Intercept (Œ≤‚ÇÄ)** | The value of **y when x = 0**. It‚Äôs where the line crosses the y-axis. Think of it as the "baseline" prediction. |
| **Slope (Œ≤‚ÇÅ)**     | The **rate of change** ‚Äî how much **y increases or decreases** for every one unit increase in **x**. |


Think of this equation like:

> üõ£Ô∏è **Your exam score = starting score + (hours studied √ó improvement per hour)**

- The **starting score** is the **intercept** ‚Äî your guess if someone didn‚Äôt study at all.
- The **improvement per hour** is the **slope** ‚Äî how much better someone scores for each additional hour of study.

Together, they form a straight line that helps us **predict** outcomes.

---

#### 3.4 - üéØ Real-World Idea:

You‚Äôre training a model to answer:

> ‚ÄúIf someone studies for 5 hours, what score will they likely get?‚Äù

This is the simplest form of regression ‚Äî **a straight line** that tries to fit your data points.

---


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)

# Predict
y_pred = model.predict(X)
```

```python
# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.title("Linear Regression with 1 Feature")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

#### 3.5 - üìå What Happened?

- We generated random points around a line (`y = 4 + 3x`)
- LinearRegression learned to fit a straight line through the data
- Red line shows the **predicted trend** (line of best fit)

### Step 4: üß™ Part 2: Linear Regression with 2 Features (Plane)

Now we have **two input features** and **one target**.  
We want to predict a value based on **two numbers** instead of one.

---

#### 4.1 - ‚úÖ What Does That Mean?

You‚Äôre trying to learn a relationship like this:

> üìä ‚ÄúIf I know **$X_1$** and **$X_2$**, can I guess **Y**?‚Äù

For example:

- **$X_1$** = Hours studied  
- **$X_2$** = Hours slept  
- **Y** = Exam score  

---

#### 4.2 - üìê Mathematically:

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2
$$

Where:
- $ x_1, x_2 $ are the input features  
- $ y $ is the predicted output  
- $ \beta_0 $ is the intercept (starting point when both inputs are 0)  
- $ \beta_1 $ and $ \beta_2 $ are the slopes for each input feature

---

#### 4.3 - üîç What's the Difference Between Each Term?

| Term        | Meaning |
|-------------|---------|
| **Intercept (Œ≤‚ÇÄ)** | The prediction when both `x‚ÇÅ` and `x‚ÇÇ` are 0. It's the "baseline" starting point of the plane. |
| **Slope (Œ≤‚ÇÅ)**     | How much **y** changes when `x‚ÇÅ` increases by 1 (while keeping `x‚ÇÇ` constant). |
| **Slope (Œ≤‚ÇÇ)**     | How much **y** changes when `x‚ÇÇ` increases by 1 (while keeping `x‚ÇÅ` constant). |

---

#### 4.4 - üß† Real-Life Analogy

> üéØ **Exam score = starting score + (study hours √ó gain) + (sleep hours √ó boost)**

- If someone **studies more**, the score should increase based on the **study slope**.
- If they **sleep more**, their performance might also improve ‚Äî that‚Äôs the **sleep slope**.

Each slope shows how that individual factor affects the prediction, while the intercept gives you a baseline.

---

#### 4.5 - üß≠ What Does It Look Like?

Instead of a line on a flat 2D chart, this model fits a **flat plane** in 3D space.

- The plane is your **best guess** surface
- Each point (dot) represents real data (input features and target)
- The model tries to position the plane so it‚Äôs **as close as possible** to all the points

---

#### 4.6 - üìå Goal of the Model

To answer:

> ‚ÄúIf someone studies 5 hours **and** sleeps 8 hours, what exam score will they likely get?‚Äù

This is how we extend linear regression into multiple inputs ‚Äî one slope per feature ‚Äî while still keeping the model **linear**.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate synthetic data
np.random.seed(1)
X = 2 * np.random.rand(100, 2)
y = 3 + 2 * X[:, 0] + 4 * X[:, 1] + np.random.randn(100)
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Plot the 3D plane
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Actual Data')
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Target")

# Plot the predicted plane
x0_grid, x1_grid = np.meshgrid(np.linspace(0, 2, 10), np.linspace(0, 2, 10))
y_grid = model.intercept_ + model.coef_[0]*x0_grid + model.coef_[1]*x1_grid
ax.plot_surface(x0_grid, x1_grid, y_grid, alpha=0.5, color='red')
plt.title("Linear Regression with 2 Features (3D Plane)")
plt.legend()
plt.show()
```

#### 4.7 - üß† Interpretation:

- The model fits a **plane** through 3D space
- You can think of this as a flat surface best explaining the relationship between 2 inputs and 1 output

### Step 7: üìê 4. What is Linear Regression?

Linear regression tries to fit a **straight line** to the data:
$$
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon
$$
Where:
- $ y $ is the target
- $ x_i $ are the input features
- $ \beta_i $ are the coefficients (weights)
- $ \epsilon $ is the error term

Goal: Find $ \beta $ values that minimize the **mean squared error**:

$$
MSE = \frac{1}{n} \sum (y_i - \hat{y}_i)^2
$$

### Step 10: üìä 7. Evaluation Metrics for Regression

Once we train a regression model, we need to measure **how good its predictions are**.  
Here are some common metrics used in regression ‚Äî explained simply:

---

#### 10.1 - 1. **Mean Absolute Error (MAE)**

$$
\text{MAE} = \frac{1}{n} \sum |y_{\text{actual}} - y_{\text{predicted}}|
$$

- Measures the **average size of the errors** (ignores whether they‚Äôre too high or too low)
- **Easy to understand**: it‚Äôs just ‚Äúhow far off‚Äù the model is, on average

> üß† Lower is better. Zero means perfect prediction.

---

#### 10.2 - 2. **Mean Squared Error (MSE)**

$$
\text{MSE} = \frac{1}{n} \sum (y_{\text{actual}} - y_{\text{predicted}})^2
$$

- Like MAE, but it **squares the errors** ‚Äî so big mistakes are penalized more
- More sensitive to outliers

> üéØ Used in gradient descent (because it's easier to compute the derivative)

---

#### 10.3 - 3. **Root Mean Squared Error (RMSE)**

$$
\text{RMSE} = \sqrt{\text{MSE}}
$$

- Just the **square root of MSE**
- Puts error back into the same **unit** as the target (e.g. dollars, prices, etc.)

> ‚úÖ Good for reporting real-world prediction quality

---

#### 10.4 - 4. **R¬≤ Score (R-squared)**

$$
R^2 = 1 - \frac{\text{Model Error}}{\text{Error of Simple Mean}}
$$

- Measures how well your model explains the variation in the data
- **1.0** means perfect prediction, **0** means your model is no better than guessing the average

> ü§ì Think of it as: ‚ÄúHow much better are we than just predicting the mean?‚Äù

---

#### 10.5 - üß† Summary Table

| Metric | What it Tells You           | Lower is Better? | Notes                              |
|--------|-----------------------------|------------------|------------------------------------|
| MAE    | Average error               | ‚úÖ Yes           | Easy to interpret                  |
| MSE    | Squared error               | ‚úÖ Yes           | Penalizes big errors more          |
| RMSE   | Error in original units     | ‚úÖ Yes           | Good for real-world interpretation |
| R¬≤     | How much variance is explained | ‚ùå Higher is better | Max = 1.0                         |

---

> ‚úÖ Tip: Use **multiple metrics together** to fully understand your model‚Äôs performance.

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R¬≤ Score:", r2)
```

### Step 11: üé• Visualizing Gradient Descent (Line Optimization)

To truly understand how **linear regression learns**, we‚Äôll simulate and visualize **gradient descent** optimizing the line to fit our data.

We'll:
- Start with random weights
- Iteratively update them using gradient descent
- Watch the line improve over time

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add bias term
X_b = np.c_[np.ones((100, 1)), X]

# Initialize parameters
theta = np.random.randn(2, 1)  # random initial values
learning_rate = 0.1
iterations = 8

# Gradient descent step (for visualization)
theta_path = [theta.copy()]
mse_vals = []
for i in range(iterations):
    gradients = 2/100 * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients
    theta_path.append(theta.copy())

# Plotting regression lines with MSE titles
plt.figure(figsize=(14, 10))
for i, theta in enumerate(theta_path):
    y_pred = X_b.dot(theta)
    mse = np.mean((y - y_pred) ** 2)
    mse_vals.append(mse)
    plt.subplot(3, 3, i+1)
    plt.scatter(X, y, label="Data", alpha=0.6)
    plt.plot(X, y_pred, color='red')
    plt.title(f"Step {i} - MSE: {mse:.2f}")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.tight_layout()

plt.suptitle("Gradient Descent: Line Fitting Over Steps", fontsize=16, y=1.02)
plt.show()
```

#### 11.1 - Can MSE Get Worse During Gradient Descent?

Usually, **no** ‚Äî if everything is set up properly.

- Gradient Descent is designed to **reduce the error** (MSE) at each step.
- For simple linear problems (like ours), the MSE usually **gets better with each step**.

---

#### 11.2 - ‚ö†Ô∏è But MSE Can Get Worse If:

1. **Learning rate is too high**  
   ‚Üí The steps are too big, and it may **overshoot** the best line.

2. **Data is noisy or complex**  
   ‚Üí Gradient descent may jump around or struggle to find the best fit.

---

#### 11.3 - ‚úÖ In This Notebook:

- We use clean, simple data
- We use a good learning rate (`0.1`)

So the **MSE should go down step by step** üëç

```python
# üìà MSE over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(len(mse_vals)), mse_vals, marker='o')
plt.title("üìâ MSE Decrease Over Gradient Descent Steps")
plt.xlabel("Step")
plt.ylabel("Mean Squared Error (MSE)")
plt.grid(True)
plt.show()
```
